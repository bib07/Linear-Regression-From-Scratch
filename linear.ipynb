```python
#linear regression try to find the best line that fits the data so that error function is minimized(using gradient descent).
# line --> y=mx+b
# Error function(mean squared error) --> E=1/n(sum(yi-y))^2
# we want to calculate lowest possible E for our line
# we can influence only m and b to calculate lowest E
#calculate partial derivative of E w.r.t m  and b respectively.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('data.csv')
df.head()
```

```python


plt.scatter(df.X,df.Y,marker='*')
plt.show()
```

```python
def loss_function(m,b,points):
    total_error=0
    for i in range(len(df)):
        x=points.iloc[i].X
        y=points.iloc[i].Y
        total_error += (y-(m*x + b)) ** 2
    total_error/float(len(points))

def gradient_descent(m_now,b_now,points,L):
    m_gradient=0
    b_gradient=0

    n=len(points)

    for i in range(n):
        x=points.iloc[i].X
        y=points.iloc[i].Y

        m_gradient += -(2/n) *x *(y-(m_now*x + b_now))
        b_gradient += -(2/n) *(y-(m_now *x +b_now))

    m=m_now-m_gradient *L
    b=b_now-b_gradient *L
    return m,b

m=0
b=0
L=0.0001
epochs=100

for i in range(epochs):
    m,b=gradient_descent(m,b,df, L)
print(m,b)

plt.scatter(df.X,df.Y,color='black')
plt.plot(list(range(20,80)),[m*x+b for x in range(20,80)],color='red')
plt.show()



```
